##Importing Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

##Data Collection & Preprocessing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
df = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
df.head()
```

```python
df['lebal']= breast_cancer_dataset.target
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['lebal'].value_counts()
```

###0 = Malignant
###1 = Benign

```python
df.groupby('lebal').mean()
```

##Separating the Features & Target

```python
X = df.drop(columns='lebal', axis=1)
Y = df['lebal']
```

```python
 print(X)
```

```python
print(Y)
```

 ## Splitting the Data into Training Data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Standarize the Data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

## Building the Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#Setting up the Layers of Neural Networks



model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])

```

```python
# Compiling the NN

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the NN

history = model.fit(X_train_std, Y_train,validation_split=.1, epochs=10)
```

```python
#Visualizing Accuracy & Loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the Model on Test Data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

```python
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
